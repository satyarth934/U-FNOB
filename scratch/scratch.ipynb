```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
inp_path = "/global/cfs/cdirs/m1012/satyarth/Data/ensemble_simulation_runs/FDL2022_data/us-digitaltwiner-dev-features/data-sim-test-2D-1000_run4/input.npy"

inp_np = np.load(inp_path)
```

```python
sim_sample = inp_np[0]
```

```python
fig, axes = plt.subplots(nrows=11, ncols=11, figsize=(20,20))

for i in range(11):
    for j in range(11):
        axes[i,j].imshow(sim_sample[:,:,i,j], origin="lower", aspect=4)
        axes[i,j].text(
            0.5, -0.4, 
            f"{np.nanmin(sim_sample[:,:,i,j]):.8e}\n{np.nanmax(sim_sample[:,:,i,j]):.8e}",
            ha="center",
            va="center",
            transform=axes[i, j].transAxes,
        )

# fig.tight_layout()
plt.show()
```
